```python
#Importing required Libraries.
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup
from sklearn.model_selection import cross_validate
from sklearn import preprocessing ,svm
from sklearn.linear_model import LinearRegression

style.use('ggplot')
```

```python
# Input the URL of the dataset we are going to use
url = 'https://www.mhanational.org/issues/2023/mental-health-america-prevalence-data'
html = urlopen(url)
```

```python
soup = BeautifulSoup(html,'lxml')
type(soup)
```

```python
#Get the Tittle
title = soup.title
print(title)
```

```python
#Print out the text
text = soup.get_text()
print(soup.text)
```

```python
#Let's see what is in the website
table = soup.find_all('table')[2]
table
```

```python
# Let's check the website for Hyperlinks
all_links = soup.find_all('a')
for link in all_links:
    print(link.get('href'))
```

```python
# Print the first 10 rows
rows = soup.find_all('tr')
print(rows[:10])
```

```python
#Let's Create the rows from webpage
for row in rows:
    row_td = row.find_all('td')
    print(row_td)
    type(row_td)
```

```python
# Removing the html tags
str_cells = str(row_td)
clean_text = BeautifulSoup(str_cells,'lxml').get_text()
print(clean_text)
```

```python
# Alternative to regular text in order to find usable text
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean,'',str_cells))
    list_rows.append(clean2)
    print(clean2)
    type(clean2)
    
```

```python
# Let's see how the tables came out in our dataframe
df = pd.DataFrame(list_rows)
df.head(50)
```

```python
# Now let's read another csv file of Mass Shootings 2023 and save into a dataframe named as df2
df2 = pd.read_csv('Data/export-14c01ebe-6437-4bd6-8980-5ed47b94e55b.csv')
```

```python
# printing the head of df2
df2.head()
```

```python
# printing the tail of df2
df2.tail()
```

```python
#Let's drop the column Operations because it has all nan values
df2.drop(['Operations'],axis = 1,inplace = True)
df2.head()
```

```python
# Renaming the columns in df2
newdf2 = df2.rename({'# Victims Injured': 'Victims Injured', '# Victims Killed': 'Victims Killed',"# Subjects-Suspects Injured":"Suspects Injured","# Subjects-Suspects Killed":"Suspects Killed","# Subjects-Suspects Arrested":"Suspects Arrested"}, axis=1)
newdf2.head()
```

```python
# Checking the null values for each column
newdf2.isnull().sum()
```

```python
# printing the dattypes for newdf2
newdf2.dtypes
```

```python
# Converting the datatypes 
dfmass = newdf2.convert_dtypes()
dfmass.dtypes
```

```python
#changing the datatype of Incident ID
dfmass['Incident ID'] = dfmass['Incident ID'].astype(str)
```

```python
# saving our datframe into csv file named as Mental_Health 2023
dfmass.to_csv('Mass_Shootings2023.csv',index = False)
```

```python
# Grouping our data by state
groupdf = dfmass.groupby('State')
 # Let's print the first entries
# in all the groups formed.
groupdf.first()
```

```python
#Counting the  number of incidents of mass shooting occur and saving it into new column names Total Incidents
dfmass['Total Incidents'] = dfmass['Incident ID'].map(dfmass['Incident ID'].value_counts())
dfmass.head()
```

```python
#grouping and doing sum of all the values based on State
massdata = dfmass.groupby(['State'], as_index=False).sum()
massdata.head()
```

```python
#Now stripping the extra space if any in the column State
dfmental1.State = dfmental1.State.str.strip()
massdata.State = massdata.State.str.strip()
```

```python
#merging the dataframe dfmental1 and massdata for further evaluation
mergedf = pd.merge(dfmental1.assign(State=dfmental1.State.astype(str)), 
         massdata.assign(State=massdata.State.astype(str)), 
         how='left', on='State')
#printing the head of dataframe
mergedf.head()
```

```python
# dropping the rows having NaN values
mergedf = mergedf.dropna()
 
# To reset the indices
Finaldf = mergedf.reset_index(drop=True)
 
# Print the dataframe head and tail
display(Finaldf.head())
display(Finaldf.tail())
```

```python
#describing the Finaldf
Finaldf.describe()
```

### Derive Conclusion

```python
#Importing  train_test_split from sklearn.model-selection
from sklearn.model_selection import train_test_split
#Importing LinearRegression from sklearn.linear_model
from sklearn.linear_model import LinearRegression
X = Finaldf[['Percentage','Numbers','Ranks']]
y = Finaldf['Total Incidents']
X_train,X_test,y_train, y_test = train_test_split(X,y,test_size = 0.4,random_state = 42)
#Instantiating LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
#printing intercept and coefficient
print(lm.intercept_)
print(lm.coef_)
```

```python
#importing metrics from sklearn
from sklearn import metrics
y_pred = lm.predict(X_test)
#printing RMSE 
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
#Importing  train_test_split from sklearn.model-selection
from sklearn.model_selection import train_test_split
#Importing LinearRegression from sklearn.linear_model
from sklearn.linear_model import LinearRegression
X = Finaldf[['Percentage','Numbers']]
y = Finaldf['Total Incidents']
X_train,X_test,y_train, y_test = train_test_split(X,y,test_size = 0.4,random_state = 42)
#Instantiating LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
from sklearn import metrics
y_pred = lm.predict(X_test)
#printing RMSE 
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

#### Conclusion:
>* The RMSE decreased when we removed Ranks From the model
>* Error is something we want to minimize
>*Thus it is unlikely that this feature is useful for Predicting AMI and should be removed from the model.
